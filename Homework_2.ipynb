```python
def a(x):
  result = 0
  while result < 10:
    result = result + x
    if result == 5:
      return result
  return result

b = a(1)
c = a(2)

print(b)
print(c)
```

```python
def a(x):
    result = 0
    while result < 10:
        result += x
        if result == 5 or result == 10:
            print(result)

a(1)
a(2)
```

```python
def tf(value):
  return 2*value-1
test_values= [0.0,0.0011,0.25,0.5,0,54,0.75,1.0]
scaled_values = [tf(value)for value in test_values]
print("Original values:",test_values)
print("Scaled values:",scaled_values)
```

```python
def a(n):
  if n == 0 or n == 1:
    return 18
  else:
    return n + a(n-1)
print(a(3))

#a(3)會進 else 計算 3 + a(2)
#a(2) 計算 2 + a(1)。
#a(1) 返回 18
#a(2) = 2 + 18 = 20
#a(3) = 3 + 20 = 23
```
